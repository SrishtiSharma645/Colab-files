#bar graph

```python
#bar grah used for combiantion of num or categorical data
#x axis...bar graph
#y axis....bar edge
cmp=["google","amazon","ms"]
profit=[34,44,55]

import numpy as np
import matplotlib.pyplot as plt
xpos=np.arange(len(cmp))   #0,1,2
plt.xticks(xpos,cmp)         #convert xpos to cmp
plt.ylabel("y")
plt.title("bar graph")
plt.bar(xpos,profit,width=0.2)
plt.show()
```

```python
#two type of data are given....overlapping ans
cmp=["ggl","amz","ms","wipro"]
profit=[65,55,67,77]
revenue=[46,78,34,32]
xpos=np.arange(len(cmp))
plt.xticks(xpos,cmp)
plt.xlabel("company")
plt.ylabel("profit/revnue")
plt.title("bar graph")
plt.bar(xpos,profit,label="prft",width=0.2)
plt.bar(xpos,revenue,label="rvnu",width=0.2)
plt.legend()
plt.show()
```

```python
#two type of data are given....separate ans
cmp=["ggl","amz","ms","wipro"]
profit=[65,55,67,77]
revenue=[46,78,34,32]
xpos=np.arange(len(cmp))
plt.xticks(xpos,cmp)
plt.xlabel("company")
plt.ylabel("profit/revnue")
plt.title("bar graph")
plt.bar(xpos-0.2,profit,label="prft",width=0.2)
plt.bar(xpos+0.2,revenue,label="rvnu",width=0.2)
plt.legend()
plt.show()
```

```python
#plt.barh....used for graw graph in y axis
plt.yticks(xpos,cmp)
plt.xlabel("profit/revnue")
plt.ylabel("company")
plt.title("bar graph")
plt.barh(xpos-0.2,profit,label="prft",height=0.3)
plt.barh(xpos+0.2,revenue,label="rvnu",height=0.3)
plt.legend()
plt.style.use('classic')     #plt.style.available
plt.show()                   #plt.style.use('classic')
```

#boxplot

Anatomy of a Boxplot:

**Median (Q2):**
The solid line inside the box represents the median of the dataset, which divides the data into two halves.(dashed line inside box:**mean**)

**Interquartile Range (IQR):**
The box itself represents the interquartile range, IQR, which is the middle 50% of the data. It spans from the first quartile (Q1) to the third quartile (Q3).

**Whiskers:**
The whiskers extend from the edges of the box to the minimum and maximum values of the data, excluding any outliers. By default, the whiskers extend up to 1.5 times the IQR from the quartiles.

**Outliers:**
Points beyond the whiskers are considered outliers and are plotted individually as points.

```python
import matplotlib.pyplot as plt
a=[12,34,12,64,23,12,34,46,56,74,5]
plt.boxplot(a)
plt.show()
```

#pie chart

**parameters**

radius=2 # to inc size of pie

autpct=%0.2f%%

shadow= True

explode=[0, 0.2, 0, 0, 0]

startangle=4

```python
#univariate
#represent % of item
%matplotlib inline
val=[43,23,24,65,54,35]
name=["a","b","c","d","e","f"]
plt.pie(val,labels=name,autopct='%0.2f%%',explode=[0.1,0,0,0.1,0,0],textprops={'color': 'black', 'fontsize': 14})
plt.show()                     # display%.......zoom in/out
```

**saving pie chart to a file using savefig**

dpi...dots per inches(resolution)              
transparent=True(remove backgroung)

pad_inches=2

```python
val=[43,23,24,65,54,35]
name=["a","b","c","d","e","f"]
plt.pie(val,labels=name,autopct='%0.2f%%',explode=[0.1,0,0,0.1,0,0],textprops={'color': 'black', 'fontsize': 14})
plt.savefig("pie.pdf",bbox_inches="tight",dpi=540)
plt.show()
```
